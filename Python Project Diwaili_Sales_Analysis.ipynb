# Exploratory Data Analysis

## *Gender*

```python
ax = sns.countplot(x = 'Gender', data = df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
Sales_gender = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False);

sns.barplot(x='Gender', y='Amount', data=Sales_gender)
```

### *From the graphs mentioned about it is quite clear that is we have more female buyers than male buyers.*

## *Age*

```python
age_ax = sns.countplot(x='Age Group', data=df, hue='Gender') #hue add diffe

for bars in age_ax.containers:
    age_ax.bar_label(bars)
```

```python
age_wise = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x='Age Group', y='Amount', data=age_wise)
```

##  *Most of the buyers are from the age group between 26-35 females*

## State

```python
#total number of orders from 10 states

sales_state = (df.groupby(["State"], as_index=False)['Orders']
               .sum()
               .sort_values(by='Orders', ascending=False)
               .head(10))

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data=sales_state, x='State', y='Orders');
```

```python
#total amount/sales from top 10 sales

sales_state = (df.groupby(["State"], as_index=False)['Amount']
               .sum()
               .sort_values(by='Amount', ascending=False)
               .head(10))

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data=sales_state, x='State', y='Amount');
```

*From above graph we can see that unexpectedly most of the orders are from Uttar Pradesh, Maharashtra and Karnataka respectively but total sales/amount  is from UP, Karnataka and then Maharashtra*

## Marital Status

```python
ax = sns.countplot(x='Marital_Status', data=df)
sns.set(rc={'figure.figsize':(5,5)})

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_marital_status = (df.groupby(['Marital_Status', 'Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False))

sns.set(rc={'figure.figsize':(6,5)})
sns.barplot(data=sales_marital_status, x='Marital_Status', y='Amount', hue='Gender')
```

*From the above graph we can see that most of the buyers are married women and they have high purchasing power*

## Occupation

```python
sns.set(rc={'figure.figsize':(30,10)})

ax = sns.countplot(x='Occupation', data=df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
occupation_df = df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=True)

sns.set(rc={'figure.figsize':(30,10)})
sns.barplot(data = occupation_df, x = 'Occupation', y='Amount')
```

*Most the buyers in the above data set are from IT, Healthcare and Aviation Sector*

## Product Category

```python
ax = sns.countplot(x='Product_Category', data=df)

sns.set(rc={'figure.figsize':(40,30)})

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
product_category_df = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=True)


sns.set(rc={'figure.figsize':(40,30)})
sns.barplot(data=product_category_df, x='Product_Category', y='Amount');
```

*Most of the sold products were from the category Food, Clothing and Apparel and Electronics and Gadgets*
